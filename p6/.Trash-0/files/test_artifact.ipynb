```python
import mlflow, os, traceback

from mlflow.tracking import MlflowClient

os.environ["GIT_PYTHON_REFRESH"] = "quiet"

experiment_name = "0-tests"

client = MlflowClient()

exp = client.get_experiment_by_name(experiment_name)
if exp is None:
    exp_id = client.create_experiment(experiment_name)
else:
    exp_id = exp.experiment_id

open("test_art.txt","w").write("hello")
print("tracking_uri:", mlflow.get_tracking_uri())
try:
    with mlflow.start_run(experiment_id=exp_id) as run:
        print("run_id:", run.info.run_id)
        mlflow.log_artifact("test_art.txt")
        print("mlflow.get_artifact_uri():", mlflow.get_artifact_uri())
except Exception as e:
    print("client exception:", e)
    traceback.print_exc()
```

```python
# Télécharger l'artefact dans un dossier local
client.download_artifacts(run_id, "tests/test_art.txt", "./downloaded_artifacts")
```
